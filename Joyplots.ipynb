```python
df = pd.read_csv('C:\Archeological Data\CPDEM100.csv')
df.head()
```

```python
df.shape
```

```python
df2 = df.T
columns = df2.columns.tolist()
columns = columns[-1:] + columns[:-1]
df2 = df2[columns]
```

```python
df2.shape
```

```python
df2.head()
```

```python
%matplotlib inline
x_range = list(range(164))
fig, axes = joypy.joyplot(df2, x_range=x_range, ylabels=False, xlabels=False, grid=False, fill=False, background='k', linecolor="w", linewidth=0.1, legend=False, overlap=0.5, figsize=(6,5),kind="values", bins=80)
axes[-1].set_xticks(x_range);
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
for a in axes[:-1]:
    a.set_xlim([-50,214])
```

```python
 fig.savefig('CuronianSpit2.png', dpi=2000)
```
